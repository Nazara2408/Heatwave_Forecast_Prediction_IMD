```python
import pandas as pd
# Rename columns to standard datetime components
df = df.rename(columns={'YEAR': 'year', 'MN': 'month', 'DT': 'day', 'HR': 'hour'})

# Create datetime column
df['datetime'] = pd.to_datetime(df[['year', 'month', 'day', 'hour']], errors='coerce')

# Drop rows with invalid datetime
df = df.dropna(subset=['datetime'])

# Set datetime as index
df = df.set_index('datetime')

# Convert all applicable columns to numeric, coercing errors
df = df.apply(pd.to_numeric, errors='coerce')

# Resample to daily and compute mean
daily_df = df.resample('D').mean()

# Preview result
daily_df.head()


```

```python
# Drop non-useful columns (like hour, day, Date) from daily data
daily_df_cleaned = daily_df.drop(columns=['hour', 'day', 'Date'], errors='ignore')

# Optional: Reset index if you want 'datetime' as a column
daily_df_cleaned = daily_df_cleaned.reset_index()

# Display the cleaned dataframe
print(daily_df_cleaned.head())
```

```python
# After resampling and before saving:
daily_df_cleaned = daily_df.copy()

# Reset index so 'datetime' becomes a column
daily_df_cleaned = daily_df_cleaned.reset_index()

# Recalculate accurate year, month, day
daily_df_cleaned['year'] = daily_df_cleaned['datetime'].dt.year
daily_df_cleaned['month'] = daily_df_cleaned['datetime'].dt.month
daily_df_cleaned['day'] = daily_df_cleaned['datetime'].dt.day

# Reorder columns if needed (optional)
cols = ['datetime', 'year', 'month', 'day'] + [col for col in daily_df_cleaned.columns if col not in ['datetime', 'year', 'month', 'day']]
daily_df_cleaned = daily_df_cleaned[cols]

# Save and download
daily_df_cleaned.to_csv('safdarjung_daily_data1.csv', index=False)

from google.colab import files
files.download('safdarjung_daily_data1.csv')
```

```python
import pandas as pd

# Step 1: Load skipping embedded header
df2_raw = pd.read_csv('/content/drive/MyDrive/ML_work/datafiles/safdarjung_2.csv', header=None, skiprows=1, on_bad_lines='skip')

# Step 2: Assign column names (41 total)
columns = ['index', 'year', 'mn', 'dt', 'max', 'min', 'aw', 'rf', 'evp',
           'drnrf(hrs)', 'drnrf(mnts)', 'ssh', 'ts', 'n', 'fff', 'd',
           'tocsq', 'du', 'ra', 'dz', 'sn', 'sl', 'ha', 'th', 'ds', 'fg',
           'ga', 't', 'g', 'dur', 't.1', 'g.1', 'dur.1', 't.2', 'g.2',
           'dur.2', 't.3', 'g.3', 'dur.3', 'unnamed: 39', 'unnamed: 40']

df2_raw.columns = columns

# Step 3: Clean column names
df2 = df2_raw.copy()
df2.columns = df2.columns.str.strip().str.lower()
df2 = df2.rename(columns={'mn': 'month', 'dt': 'day'})

# Step 4: Handle datetime
df2 = df2.dropna(subset=['year', 'month', 'day'])
df2[['year', 'month', 'day']] = df2[['year', 'month', 'day']].astype(int)
df2['datetime'] = pd.to_datetime(df2[['year', 'month', 'day']], errors='coerce')

# Step 5: Check output
print(df2[['datetime', 'max', 'min']].head())
```

```python
df1 = pd.read_csv('/content/safdarjung_daily_data1.csv')
df1_reduced = df1[[
    'datetime',     # date
    'DBT',          # dry bulb temp
    'WBT',          # wet bulb temp
    'DPT',          # dew point
    'RH',           # relative humidity
    'SLP', 'MSLP',  # sea-level pressure
    'hour'          # hours measured per day (optional)
]].copy()


# Assuming df2 has these columns already cleaned
df2_reduced = df2[['datetime', 'max', 'min']]
```
