```python
import pandas as pd

# Load the final cleaned and merged dataset
merged_df = pd.read_csv('/content/drive/MyDrive/ML_work/datafiles/station_data/ayanagar_merged.csv')
merged_df['datetime'] = pd.to_datetime(merged_df['datetime'])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
merged_df.drop(columns=['hour'], inplace=True)
print(merged_df.head)
```

```python
# Step 2: Feature Engineering
df = merged_df.copy()

# Create new features
df['temp_diff'] = df['max'] - df['min']
df['dewpoint_diff'] = df['dbt'] - df['dpt']

# Optional: Drop datetime if not using it directly
df = df.drop(columns=['datetime'])
```

```python
# Step 3: Define Features and Target
X = df.drop(columns=['heatwave'])
y = df['heatwave']
```

```python
# Step 4: Feature Scaling (optional for tree-based models, but good practice)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Step 5: Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Step 1: Import required libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor  # <== this was missing!
import matplotlib.pyplot as plt
import seaborn as sns



# Step 6: Try a better model (Random Forest Regressor)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred = rf_model.predict(X_test)

# Evaluation
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"âœ… RÂ² Score: {r2}")
print(f"âœ… RMSE: {rmse}")
```

```python
# Step 7: Cross-validation check
cv_scores = cross_val_score(rf_model, X_scaled, y, cv=5, scoring='r2')
print(f"âœ… Cross-validated RÂ² scores: {cv_scores}")
print(f"âœ… Mean CV RÂ²: {cv_scores.mean()}")
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np

```

```python
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'DecisionTree': DecisionTreeRegressor(random_state=42),
    'RandomForest': RandomForestRegressor(n_estimators=100, random_state=42),
    'GradientBoosting': GradientBoostingRegressor(n_estimators=100, random_state=42),
    'SVR': SVR()
}
```

```python
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    cv_r2 = cross_val_score(model, X, y, cv=5, scoring='r2')

    results.append({
        'Model': name,
        'R2 Score': r2,
        'RMSE': rmse,
        'CV R2 Mean': np.mean(cv_r2)
    })

# Convert to DataFrame
results_df = pd.DataFrame(results).sort_values(by='R2 Score', ascending=False)
```
